## Handle Missing Data with replace function

```python
import pandas as pd
import numpy as np
df = pd.read_csv('data/006_weather_data.csv')
df
```

```python
new_df = df.replace(-99999, np.NaN)
new_df
```

```python
new_df = df.replace([-99999, -8888.0], np.NaN)
new_df
```

```python
new_df = df.replace({
    'temperature': -99999,
    'windspeed': -8888,
    'event': '0'
}, np.NaN)
new_df
```

```python
df = pd.read_csv('data/006_2_weather_data.csv')
df
```

```python
df_new = df.replace({
    -99999: np.NaN,
    'No Event': np.NaN,
    -8888: np.NaN
})
df_new
```

```python
df_new = df.replace({
    'temperature': '[A-Za-z]',
    'windspeed': '[A-Za-z]',
}, '', regex=True)
df_new
```
