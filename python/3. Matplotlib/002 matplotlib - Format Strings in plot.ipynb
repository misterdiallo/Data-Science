### Format Strings in plot

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=[1,2,3,4,5,6,7]
y=[50,51,52,48,47,49,46]
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y, 'g+')
```

```python
plt.plot(x,y, 'g+--')
```

```python
plt.plot(x,y, '--r*')
```

```python
plt.plot(x,y, 'g+--', markersize=40)
```

```python
plt.plot(x,y, 'g+--', alpha=0.2)
```

```python
plt.plot(x,y, 'g+--', alpha=0.4)
```
