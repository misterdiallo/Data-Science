### Axes labels, Legend, Grid

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
days=[1,2,3,4,5,6,7]
min_t=[43,42,40,44,33,35,37]
max_t=[50,51,52,48,47,49,46]
avg_t=[45,48,48,46,40,42,41]
```

```python
plt.plot(days, min_t,)
plt.plot(days, max_t)
plt.plot(days, avg_t)
```

```python
plt.title('Weather Data')
plt.xlabel('Days')
plt.ylabel('Temperatures')
plt.plot(days, min_t, label="Max")
plt.plot(days, max_t, label="Min")
plt.plot(days, avg_t, label="Avg")
plt.legend()
```

```python
plt.title('Weather Data')
plt.xlabel('Days')
plt.ylabel('Temperatures')
plt.plot(days, min_t, label="Max")
plt.plot(days, max_t, label="Min")
plt.plot(days, avg_t, label="Avg")
plt.legend(loc='best')
```

```python
plt.title('Weather Data')
plt.xlabel('Days')
plt.ylabel('Temperatures')
plt.plot(days, min_t, label="Max")
plt.plot(days, max_t, label="Min")
plt.plot(days, avg_t, label="Avg")
plt.legend(shadow=True)
```

```python
plt.title('Weather Data')
plt.xlabel('Days')
plt.ylabel('Temperatures')
plt.plot(days, min_t, label="Max")
plt.plot(days, max_t, label="Min")
plt.plot(days, avg_t, label="Avg")
plt.legend(shadow=True, fontsize="small")
```

```python
plt.title('Weather Data')
plt.xlabel('Days')
plt.ylabel('Temperatures')
plt.plot(days, min_t, label="Max")
plt.plot(days, max_t, label="Min")
plt.plot(days, avg_t, label="Avg")
plt.legend(shadow=True,)
plt.grid()
```
