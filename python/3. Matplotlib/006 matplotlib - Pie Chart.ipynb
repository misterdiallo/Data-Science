### Pie Chart

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
plt.pie(exp_vals,labels=exp_labels)
```

```python
plt.axis('equal')
plt.pie(exp_vals,labels=exp_labels)
```

```python
plt.axis('equal')
plt.pie(exp_vals,labels=exp_labels)
plt.show()
```

```python
plt.axis('equal')
plt.pie(exp_vals,labels=exp_labels, radius=1.5)
plt.show()
```

```python
plt.axis('equal')
plt.pie(exp_vals,labels=exp_labels, radius=2, autopct='%0.2f%%')
plt.show()
```

```python
plt.axis('equal')
plt.pie(exp_vals,labels=exp_labels, radius=2, autopct='%0.0f%%')
plt.show()
```

```python
plt.axis('equal')
plt.pie(exp_vals,labels=exp_labels, radius=2, autopct='%0.1f%%', explode=[0,0,0,0,0])
plt.show()
```

```python
plt.axis('equal')
plt.pie(exp_vals,labels=exp_labels, radius=2, autopct='%0.1f%%', explode=[0,0,0.3,0,0])
plt.show()
```

```python
plt.axis('equal')
plt.pie(exp_vals,labels=exp_labels, radius=2, autopct='%0.1f%%', explode=[0,0.1,0.1,0,0])
plt.show()
```

```python
plt.axis('equal')
plt.pie(
    exp_vals,
    labels=exp_labels, 
    radius=2, 
    autopct='%0.1f%%', 
    explode=[0,0.1,0.1,0,0],
    startangle=180,
)
plt.show()
```

```python
plt.axis('equal')
plt.pie(
    exp_vals,
    labels=exp_labels, 
    radius=2, 
    autopct='%0.1f%%', 
    explode=[0,0.1,0.1,0,0],
    startangle=90,
)
plt.show()
```

```python
plt.axis('equal')
plt.pie(
    exp_vals,
    labels=exp_labels, 
    radius=2, 
    autopct='%0.1f%%', 
    explode=[0,0.1,0.1,0.1,0.2],
    startangle=180,
    shadow=True,
)
plt.show()
```

```python
plt.axis('equal')
plt.pie(
    exp_vals,
    labels=exp_labels, 
    radius=2, 
    autopct='%0.1f%%', 
    explode=[0,0.1,0.1,0.1,0.2],
    startangle=180,
    shadow=True,
    counterclock=True,
)
plt.show()
```
