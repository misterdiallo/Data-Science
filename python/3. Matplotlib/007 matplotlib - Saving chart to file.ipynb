<h2 style="color:green" align="center">Saving chart to a file using savefig</h2>

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels,radius=2, autopct='%0.1f%%', explode=[0,0.1,0.1,0,0])
plt.savefig("piechart.pdf", bbox_inches="tight", pad_inches=1, transparent=True)
```

<h2>It can also save in pdf format and to a different location on computer by specifying absolute path<h2>

```python
plt.pie(exp_vals,labels=exp_labels,radius=2, autopct='%0.1f%%', explode=[0,0.1,0.1,0,0])
plt.savefig("c:/code/piechart.pdf", bbox_inches="tight", pad_inches=10, transparent=True)
```
